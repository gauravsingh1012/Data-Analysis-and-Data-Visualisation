```
#SELECT * FROM df WHERE location IS NULL and row_num = 1;

df[df['location'].isna()].head(1)
```

```
#SELECT * FROM df WHERE retweets IS not NULL and row_num = 1;

df[df['retweets'].notna()].head(1)
```

```
#SELECT hashtags, count(*) FROM df GROUP BY hashtags;

df.groupby('hashtags').size()
```

```
#Notice that in the pandas code we used size() and not count(). 
#This is because count() applies the function to each column, returning the number of not null records within each.

df.groupby('hashtags').size().count()
```

```
#Alternatively, we could have applied the count() method to an individual column:

df.groupby('hashtags')['replies'].count()
```

```
#Multiple functions can also be applied at once. For instance, say we’d like to 
#see how retweets amount differs by hashtag - agg() allows you to pass a dictionary to your grouped DataFrame, 
#indicating which functions to apply to specific columns.
#SELECT hashtags, sum(retweets), COUNT(*) FROM df GROUP BY hashtags;
import numpy as np
df.groupby('hashtags').agg({'retweets': np.sum, 'hashtags': np.size})

```

```
df['date'] = pd.to_datetime(df['date'])
df['year'], df['month'] = df['date'].dt.year, df['date'].dt.month

#The groupby() function returns a GroupBy object, but essentially describes how the rows of the original data set has been split. 
#the GroupBy object .groups variable is a dictionary whose keys are the computed unique groups and corresponding values being the 
#axis labels belonging to each group. For example:
df.groupby(['year']).groups.keys()
```

```
df.index
```

```
#SELECT year, sum(retweets),sum(likes) FROM df GROUP BY year;
df.groupby('year')['retweets','likes'].sum()
```

```
#SELECT hashtags,sum(retweets),sum(likes) from df GROUP by hashtags;
df.groupby('hashtags')['retweets','likes'].sum()
```

```
#SELECT hashtags,mean(retweets),mean(likes) from df GROUP by year,hashtags;
df.groupby(['year','hashtags'])['retweets','likes'].mean()
```

```
#select year, hashtag, count(*) from df;
df.groupby(['year','hashtags']).size()
```
